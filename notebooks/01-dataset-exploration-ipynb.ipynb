```python
import os

base_path = "/kaggle/input/datasets/tawsifurrahman/covid19-radiography-database/COVID-19_Radiography_Dataset"

for category in ["COVID", "Normal", "Lung_Opacity", "Viral Pneumonia"]:
    
    image_folder = os.path.join(base_path, category, "images")
    
    print(category, len(os.listdir(image_folder)))
```

```python
import matplotlib.pyplot as plt
from PIL import Image

classes = ["COVID", "Normal", "Lung_Opacity", "Viral Pneumonia"]

plt.figure(figsize=(12,3))

for i, cls in enumerate(classes):
    img_folder = os.path.join(base_path, cls, "images")
    img_name = os.listdir(img_folder)[0]

    img = Image.open(os.path.join(img_folder, img_name))

    plt.subplot(1,4,i+1)
    plt.imshow(img, cmap="gray")
    plt.title(cls)
    plt.axis("off")

plt.show()
```
